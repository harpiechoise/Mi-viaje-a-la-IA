### ¿Que es matplotlib?

Es una **libreria** de python que es util para ya no ver dataframes si no una representacion real de los datos, ¿esto es util para?, cuando nos enfrentamos a problemas de regresion o clusterizacion de datos, debemos ver que es lo que esta clasificando nuestro programa, y aun mas atras antes de comenzar que predictor usar para el problema representado se recomienda ver la representacion de los datos, y al igual podemos tabular y representar funciones lineales de esta manera bueno vamos a ello

al importar matplotlib debemos importar la clase pyplot de la libreria porque es la que contiene los graficos, cuando trabajamos con IPython o **jupyter notebook** este ultimo nos ofrece un comando de celda magica el cual es %matplotlib inline, para ver los graficos en el mismo programa por lo que debemos ejecutar ese comando cuando trabajemos con **jupyter notebook** 

##### Graficos de linea

Ahora veremos como hacer los graficos de linea **(line charts)** de matplotlib

```python
#Importando matplotlib
import matplotlib.pyplot as plt
import numpy as np
#Ejecutando el comando para jupyter
%matplotlib inline
```

```python
x = np.linspace(0, 10, 10)
```

El metodo de **Numpy**, **linespace** es otra forma de generar datos que forma una linea recibiendo 3 argumentos
1. Punto inicial (En este caso 0)
2. Punto Final(En este caso 10)
3. La cantidad de puntos entre el punto inicial y el final (en este caso 10)

```python
y = np.sin(x)
```

Ahora vamos a crear una onda sinusoidal aplicando el metodo **sin** de **numpy** a todos los 10 puntos de x es decir en el eje de las **x** tendremos una linea, y el de las **y** una el seno de las mismas por que dara por resultado una onda  

**Nota** 
La funcion **numpy.sin** es una operacion elemento por elemento lo que significa esque sacara el seno de todos los elementos de la matriz x

ahora para verlo en un grafico solo debemos ejecutar el metodo **plt.plot()** que recive como primer argumento el eje de las x y como segundo el eje de las y de **matplotlib** y finalmente **plt.show()** del mismo

cada vez que queramos visualizar un grafico, devemos llamar al metodo **show()**

```python
plt.plot(x,y)
plt.show()
```

Otra cosa que podemos hacer es agregarle etiquetas de texto a ambos ejes  con los siguientes comandos

~~~python
matplotlib.pyplot.xlabel('Etiqueta')
matplotlib.pyplot.ylabel('Etiqueta')
#En este caso matplotlib.pyplot lo importamos como plt
~~~

y tambien darle un titulo con el metodo title
~~~python
matplotlib.pyplot.title('Titulo')
#En este caso matplotlib.pyplot lo importamos como plt
~~~

Vamos a ello

```python
plt.plot(x,y)
plt.xlabel('Tiempo');
plt.ylabel('Alguna funcion')
plt.title('Mi grafico de cualquier cosa')
plt.show()
```

Otra cosa a notar esque cada vez que llamamos al metodo **show()** nuestro grafico se pone en blanco denuevo y si quieremos graficar por segunda vez devemos definir denuevo nuestras variables de las x y las y

Algo que podemos notar esque el grafico no sale como una onda, si no como un monton de lineas conectadas, esto se debe a la densidad de puntos, y las cantidad de pixeles de la pantalla ya que ese es el medio de salida del grafico por lo que estamos pidiendo graficar una densidad de 10 puntos entre el origen y la salida, al ser una cantidad infima de informacion se ve así

Ahora hagamos el mismo ejemplo pero con mas puntos

```python
X = np.linspace(0,10,100)
Y = np.sin(X)
plt.plot(X,Y)
plt.xlabel('Tiempo')
plt.ylabel('Sin(x)')
plt.title('Mi grafico mejorado')
```

### Graficos de dispersion

Vamos a ver ahora otra popular forma de visualizacion de datos llamada, grafico de dispersion, que dentro del machine learning es una de las mas usadas, pero al final terminas usando muchas mas.

En el ejemplo pasado vimos solo una funcion, y dentro del machine learning no aplicamos una sola funcion, si no que pasamos una cantidad grande de datos, sin saber la funcion exacta de la que estos datos fueron generados, en lo que realmente se basa el **machine learning** es aproximacion de funciones, y uno de los problemas mas comunes es saber que modelo se ajusta mejor a los datos, a travez de la visualizacion de los mismos

Ahora vamos a cargar los datos de regresion lineal, para ver como se deben graficar y visualizar

Comenzaremos leyendo el archivo csv como ya sabemos, con pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
M = pd.read_csv('Data/data_1d.csv',header=None).as_matrix()
# Solo importe dos columnas para mayor comodidad al explicar
```

```python
#El eje de las X es la primera columna y el de las Y la segunda columna
X = M[:,0] # [:,n] significa que seleccione todos los datos de la dimension n
Y = M[:,1] # [n] Significa la fila pero :,n significa la columna
```

```python
plt.scatter(X,Y)
plt.show()
```

Si queremos hacer dos graficos superpuestos, solo debemos hacer un plot de todos los datos que queremos superponer, ya que como dijimos la vez pasada, cada vez que llamamos al metodo **show()** el grafico se limpia para evitar la superposicion de graficos, que no queremos superponer, pero podemos superponer cualquier cantidad de graficos antes del metodo **show**

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1
```

```python
plt.scatter(X,Y)
plt.plot(x_line,y_line)
plt.show()
```

Lo que logramos es modificar la desviacion estandar de un vector aleatorio uniforme y para entenderlo mejor vamos a visualizarlo con el grafico de barras. vamos con un ejemplo unidimensional

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
r = np.random.randn(10000)
plt.hist(r, bins=100)
plt.show()
```

##### Sampleando desde una distribuicion esferica y alineada en sus ejes elipticamente

En machine learning siempre tendremos que trabajar con datos multidimencionales, supongamos que queremos samplear una distribucion 2-D con un promedio 0 y una variante, esto se llama distribucion gaussiana esferica, porque cada dimension es independiente y no coorelacionada a la otra, pero tienen variantes parecidas, cuando graficamos los datos, se dispersan en forma circular, por supuesto podemos usar la misma funcion de antes para cada dimension

```python
r = np.random.randn(10000,2)
```

Ahora vamos a corrovorar si los puntos estan distribuidos en forma circular con un grafico de dispersion

```python
plt.scatter(r[:,0],r[:,1])
plt.show()
```

Ahora supongamos que queremos una distribucion eliptica, esto quiere decir que las variantes son distintas para cada dimension, supongamos que la desviacion estandar para la segunda dimension quiero que sea 5 y el promedio sea 2 

```python
r[:,1] = 5*r[:,1] + 2
```

Volvere a hacer el grafico para comprobar los datos

```python
plt.scatter(r[:,0],r[:,1])
plt.show()
```

Podemos notar que aun se ve como un circulo pero eso es porque matplotlib esta escalando los ejes hacia la direccion de donde estan balanceados los datos para arreglarlo devemos especificar que los ejes sean iguales

```python
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```

Ahora estos son los datos con los ejes igualados en ese punto pensaras es posible crear elipses que no esten alineados con los ejes, esto quiere decir que tenemos que tener graficos donde sus dimensiones esten coorelacionadas con las otras y esto se llama distribucion general multivariante, para eso vamos a definir una matriz covariante

```python
cov = np.array([[1,0.8],[0.8,3]]) # la covarianza entre ambas dimensiones es 0.8

#Ahora vamos a importar el metodo para distribucion normal multivariante
from scipy.stats import multivariate_normal as mvn
```

```python
mu = np.array([0,2]) # este es el promedio 
```

```python
r = mvn.rvs(mean=mu, cov=cov, size=1000) # Size se corresponde a cuantos puntos se generan
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```

Como podemos notar ya hay un poco de coorelacion, tambien esta distribucion se puede lograr con numpy

```python
r = np.random.multivariate_normal(mean=mu,cov=cov,size=10000)
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```

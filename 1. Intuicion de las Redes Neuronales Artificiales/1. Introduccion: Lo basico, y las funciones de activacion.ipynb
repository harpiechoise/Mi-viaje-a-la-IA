#### Funcion de activacion

Ahora pasaremos a ver distintos tipos de funciones de activacion para finalmente saber cual escoger en cada caso

**Activacion umbral**

![Activacion Umbral](http://chem-eng.utoronto.ca/~datamining/dmc/images/ANN_Unit_step.png)

lo que hace es devolver 1 si la entrada X es mayor o igual a 0 y 0 si la entrada X es menor o igual a 0 esta es una funcion bastante simple.
\begin{equation}
f_{(x)} = 
\begin{cases}
    1 & \text{if}\ x \geq 0 \\
    0 & \text{if}\ x < 0
\end{cases}
\end{equation}

**Donde x es el valor de la suma de los pesos**

**Sigmoide**

![Sigmoide](https://upload.wikimedia.org/wikipedia/commons/thumb/6/66/Funci%C3%B3n_sigmoide_01.svg/300px-Funci%C3%B3n_sigmoide_01.svg.png)

Cuya formula es la siguiente

\begin{equation}
\theta_{(x)} = \frac{1}{1 + e^{-x}}
\end{equation}

**Donde X es el valor de la suma de los pesos**

Esta funcion se usa generalmente para problemas de **Regresion lineal** lo bueno de esta funcion es que es mas suave no varia tanto a diferencia de la anterior no es tan radical es una progresion gradual lo menor a cero se va hacia abajo y gradualmente aumenta el valor, esta funcion es muy util en la **ultima capa** especialmente cuando estas tratando de predecir probabilidades 

**Funcion Rectificadora**

[ReLu](https://qph.fs.quoracdn.net/main-qimg-b0a1423dbff251a5d46117cc72943d2b)

\begin{equation}
f_{(x)} = 
\begin{cases}
    x & \text{if}\ x \geq 0 \\
    0 & \text{if}\ x < 0
\end{cases}
\end{equation}


Esta funcion es la mas popular dentro del **Deep Learning** ya que da 0 si la entrada es menor a 0 y si no tiene un aumento gradual, vamos a ver entrados en materia en los casos que se ocupa, pero hay que recordar que es la funcion de activacion mas popular en el **Deep Learning** ya que es la que mas se asemeja a como funcionan las redes neuronales humanas

**Tangente Hyperbolica**

![Tahn](https://upload.wikimedia.org/wikipedia/commons/thumb/8/87/Hyperbolic_Tangent.svg/1200px-Hyperbolic_Tangent.svg.png)

**Formula**

\begin{equation}
f_{(x)} = \frac{1-e^{-2x}}{1+e^{-2x}}
\end{equation}

Es muy similar a la sigmoide pero los valores van por debajo de cero en este caso desde -1 a 1, no vamos a ir tan en profundidad con estas funciones solo explicaremos lo dicho y mas que se vea como estas funciones

**Como utilizar estas funciones**

¿Teniendo una salida de 0 y 1 es decir una salida binaria que funcion de activacion debemos aplicar?

viendo las imagenes tenemos 2 opciones, la funcion **Sigmoide** que podria calzar ya que sus valores van de 0 a 1 y la otra funcion es la de **umbral** que nos dara 0 o 1 que igual calza en el perfil del problema, pero aun asi la funcion **Sigmoide** devolveria la probabilidad de 0 o 1 y la funcion **Umbral** nos daria el 0 o 1 sin probabilidad.
